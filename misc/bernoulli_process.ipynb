## Bayes theorem

\begin{align}
P(A|B) &= \frac{P(B|A)\cdot P(A)}{P(B)} \\
\text{posterior}&= \frac{\text{likelihood}\cdot \text{prior}}{\text{evidence}}
\end{align}

### likelihood = bernoulli process



```python
#| code-summary: "import libraries"
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
params = np.empty((0, 3))
np.random.seed(1)
did_it_rain = np.random.binomial(
    n=1,     # number of trials. n=1 is a bernoulli trial
    p=0.5,   # probability of success
    size=7*52  # number of draws
)

x = np.linspace(0, 1, 100)
params = np.append(params, [[1, 10, 1/11]], axis=0)
for i in range(len(did_it_rain)):
    a = params[-1,0]+did_it_rain[i]
    b = params[-1,1]+(1-did_it_rain[i])
    params = np.append(
        params,
        [[
            a,
            b,
            a/(a+b)
        ]], axis=0)
# a = 2
# b = 9
# y1 = stats.beta.pdf(x, params[-1,0], params[-1,1])

# did_it_rain = np.random.binomial(1, 0.5, 100)

# y2 = stats.beta.pdf(x, 2, 11)
fig, ax = plt.subplots()
ax.plot(x, y1)
# ax.plot(x, y2)
```
